```python
import pandas as pd
import numpy as np
import yfinance as yf
from datetime import datetime, timedelta
import pytz

# =========================
# Configuración
# =========================
TICKERS = {
    "Nasdaq Composite": "^IXIC",
    "S&P 500": "^GSPC",
    "BMV IPC": "^MXX",
    "Russell 2000": "^RUT",
    "Dow Jones": "^DJI",
    "USD/MXN (spot)": "MXN=X"
}

# Zona horaria del usuario (CDMX)
TZ_USER = pytz.timezone("America/Mexico_City")

# Zona horaria por mercado (para definir "hoy" intradía)
EXCHANGE_TZ = {
    "^IXIC": "America/New_York",
    "^GSPC": "America/New_York",
    "^RUT":  "America/New_York",
    "^DJI":  "America/New_York",
    "^MXX":  "America/Mexico_City",
    "MXN=X": "America/Mexico_City",
}

# =========================
# Funciones auxiliares
# =========================
def get_exchange_tz(ticker: str):
    name = EXCHANGE_TZ.get(ticker, "America/New_York")
    return pytz.timezone(name)

def normaliza_cols(df: pd.DataFrame) -> pd.DataFrame:
    if isinstance(df.columns, pd.MultiIndex):
        df.columns = [c[0] for c in df.columns]
    return df

def limpia_ceros(df: pd.DataFrame, cols=("Open", "Close", "Adj Close")) -> pd.DataFrame:
    df = df.copy()
    for c in cols:
        if c in df.columns:
            df.loc[df[c].fillna(0) == 0, c] = np.nan
    return df

def descarga_hist_diario(ticker: str, dias: int = 180) -> pd.DataFrame:
    df = yf.download(
        ticker, period=f"{dias}d", interval="1d",
        progress=False, auto_adjust=False, group_by="column", threads=False
    )
    df = normaliza_cols(df)
    df = limpia_ceros(df)
    return df

def descarga_intradia(ticker: str, dias: int = 7, interval="5m") -> pd.DataFrame:
    df = yf.download(
        ticker, period=f"{dias}d", interval=interval,
        progress=False, auto_adjust=False, prepost=False, group_by="column", threads=False
    )
    df = normaliza_cols(df)
    if df.empty:
        return df
    if df.index.tz is None:
        df.index = df.index.tz_localize("UTC")
    df = limpia_ceros(df)
    return df

def ultimo_close_habil_anterior(df_d: pd.DataFrame) -> float | None:
    """
    Regresa el penúltimo cierre diario válido (cierre hábil anterior).
    """
    if df_d is None or df_d.empty:
        return None
    col = "Close" if "Close" in df_d.columns else ("Adj Close" if "Adj Close" in df_d.columns else None)
    if col is None:
        return None
    closes = df_d[col].dropna()
    if closes.shape[0] < 2:
        return None
    return float(closes.iloc[-2])

def close_ultimo_diario(df_d: pd.DataFrame) -> float | None:
    """
    Último cierre diario disponible (hoy si ya cerró).
    """
    if df_d is None or df_d.empty:
        return None
    col = "Close" if "Close" in df_d.columns else ("Adj Close" if "Adj Close" in df_d.columns else None)
    if col is None:
        return None
    closes = df_d[col].dropna()
    if closes.empty:
        return None
    return float(closes.iloc[-1])

def apertura_y_ultimo_intradia(df_i: pd.DataFrame, tz_ex):
    """
    Regresa apertura intradía de hoy (primer Close del día)
    y último precio intradía disponible.
    """
    if df_i is None or df_i.empty or "Close" not in df_i.columns:
        return None, None

    data = df_i.copy()
    if data.index.tz is None:
        data.index = data.index.tz_localize("UTC")
    data = data.tz_convert(tz_ex)

    hoy_ex = datetime.now(tz_ex).date()
    df_hoy = data[data.index.date == hoy_ex]

    # último precio intradía global
    last_price = float(data["Close"].dropna().iloc[-1]) if not data["Close"].dropna().empty else None

    if df_hoy.empty:
        return None, last_price

    # apertura estimada = primer Close de hoy
    first_close = df_hoy["Close"].dropna()
    open_today = float(first_close.iloc[0]) if not first_close.empty else None

    return open_today, last_price

def obtener_metricas(ticker: str):
    tz_ex = get_exchange_tz(ticker)

    # Diario
    df_d = descarga_hist_diario(ticker, dias=180)
    close_ayer = ultimo_close_habil_anterior(df_d)
    close_ult_diario = close_ultimo_diario(df_d)

    # Intradía
    df_i = descarga_intradia(ticker, dias=7, interval="5m")
    open_hoy, last_price = apertura_y_ultimo_intradia(df_i, tz_ex)

    # Si no hay intradía, usamos el último diario disponible
    if last_price is None:
        last_price = close_ult_diario

    return open_hoy, close_ayer, last_price

def formato_num(row, fx_name="USD/MXN (spot)"):
    is_fx = (row.name == fx_name)
    dec = 4 if is_fx else 2

    for c in ["Apertura", "Cierre (ayer)", "Último", "% Hoy"]:
        if c in row.index and pd.notna(row[c]):
            if c == "% Hoy":
                row[c] = round(row[c], 2)
            else:
                row[c] = round(row[c], dec)
    return row

# =========================
# Pipeline
# =========================
registros, avisos = [], []

for nombre, tic in TICKERS.items():
    try:
        open_hoy, close_ayer, last_price = obtener_metricas(tic)

        if close_ayer is None or close_ayer == 0:
            avisos.append(f"{nombre} ({tic}): sin cierre hábil anterior válido")
            continue

        # % Hoy real = último vs cierre hábil anterior
        hoy_abs = (last_price - close_ayer) if (last_price is not None) else np.nan
        hoy_pct = (hoy_abs / close_ayer * 100.0) if pd.notna(hoy_abs) else np.nan

        registros.append({
            "Instrumento": nombre,
            "Ticker": tic,
            "Apertura": open_hoy,
            "Cierre (ayer)": close_ayer,
            "Último": last_price,
            "% Hoy": hoy_pct,
        })

    except Exception as e:
        avisos.append(f"{nombre} ({tic}): {type(e).__name__}: {e}")

df = pd.DataFrame(registros).set_index("Instrumento")

# Orden deseado
orden = [
    "Nasdaq Composite", "S&P 500", "BMV IPC",
    "Russell 2000", "Dow Jones", "USD/MXN (spot)"
]
df = df.reindex([i for i in orden if i in df.index])

if not df.empty:
    df = df.apply(formato_num, axis=1)

from IPython.display import display
df_print = df.copy()

# Formato de % Hoy (string con %)
if "% Hoy" in df_print.columns:
    df_print["% Hoy"] = df_print["% Hoy"].map(lambda x: f"{x:.2f}%" if pd.notna(x) else "—")

display(df_print)

# =========================
# LaTeX BONITO desde Python
# =========================

# 1) Orden de columnas para PDF
cols_order = ["Ticker", "Apertura", "Cierre (ayer)", "Último", "% Hoy"]
df_latex = df_print.copy()[cols_order]

# 2) Cuerpo LaTeX limpio:
latex_body = df_latex.to_latex(
    escape=True,
    index=True,          # sí queremos Instrumento como primera columna
    index_names=False,   # NO imprime la palabra "Instrumento"
    bold_rows=False,     # sin negritas automáticas
    header=True,
    column_format="lrrrrr"  # Instrumento + 5 cols numéricas
)

# 3) Envolver con estilo booktabs y espaciado
latex_table = r"""
\begin{table}[htbp]
\centering
\footnotesize
\caption{Resumen de mercados: apertura, cierre hábil anterior y variación de hoy}
\label{tab:resumen_mercados}
\setlength{\tabcolsep}{6pt}
\renewcommand{\arraystretch}{1.2}
\resizebox{\textwidth}{!}{%%
%s
}
\end{table}
""" % (latex_body)

print("%% --- Copia y pega en tu documento LaTeX ---")
print(latex_table)
print("%% --- Fin del bloque LaTeX ---")

with open("/content/resumen_mercados.tex", "w", encoding="utf-8") as f:
    f.write(latex_table)

print("\nArchivo guardado en Colab: /content/resumen_mercados.tex")
print("Actualizado (CDMX):", datetime.now(TZ_USER).strftime("%Y-%m-%d %H:%M %Z"))

if avisos:
    print("\nAvisos:")
    for a in avisos:
        print("-", a)
```

```python
# ============================================
# Best & Worst del S&P/BMV IPC -> Tablas LaTeX
# Google Colab | Yahoo Finance (yfinance)
# Ranking por rendimiento diario (% vs cierre previo)
# Versón robusta con guardas anti NaN/IndexError
# ============================================

!pip -q install yfinance pandas pytz

import pandas as pd
import numpy as np
import yfinance as yf
from datetime import datetime
import pytz
import os
import warnings
warnings.filterwarnings("ignore")

# --- Configuración ---
TIMEZONE = "America/Mexico_City"
CSV_PATH = "/content/ipc_tickers.csv"  # opcional; sube tu CSV con columna 'ticker'

# Lista por defecto (ajústala a tu composición / símbolos Yahoo).
DEFAULT_TICKERS = [
    "AMXL.MX","WALMEX.MX","GFNORTEO.MX","CEMEXCPO.MX","FEMSAUBD.MX","GMEXICOB.MX",
    "BIMBOA.MX","ALFAA.MX","TLEVISACPO.MX","KIMBERA.MX","BBAJIOO.MX","ASURB.MX",
    "GAPB.MX","OMAB.MX","MEGACPO.MX","PINFRA.MX","ALSEA.MX","GENTERA.MX","ORBIA.MX","PENOLES.MX"
    # Si quieres que salgan emisoras como VOLAR, TERRA13, GCARSO, Q, etc.,
    # agrega aquí su ticker exacto de Yahoo (ej. Q.MX, TERRA13.MX, GCARSOA1.MX, etc.)
]

# --- Cargar tickers ---
def load_tickers():
    try:
        df = pd.read_csv(CSV_PATH)
        assert "ticker" in df.columns
        tks = sorted(set(df["ticker"].dropna().astype(str)))
        print(f"Usando {len(tks)} tickers del CSV:", tks[:10], "...")
        return tks
    except Exception:
        tks = sorted(set(DEFAULT_TICKERS))
        print("No se encontró CSV válido. Usando lista por defecto.")
        print(f"{len(tks)} tickers:", tks[:10], "...")
        return tks

TICKERS = load_tickers()
tz = pytz.timezone(TIMEZONE)
today_mx = datetime.now(tz).date()

# --- Utilidades ---
def latex_escape(s: str) -> str:
    if not isinstance(s, str):
        return s
    return (s.replace("\\", r"\\")
             .replace("&", r"\&")
             .replace("_", r"\_")
             .replace("%", r"\%")
             .replace("#", r"\#")
             .replace("{", r"\{")
             .replace("}", r"\}")
             .replace("$", r"\$"))

def get_daily_returns(tickers):
    """
    Descarga diarios y calcula rendimiento del día:
    (close_hoy / close_ayer - 1) * 100
    Esto replica 'Rendimientos del día por emisora' como en la imagen.
    """
    if not tickers:
        return pd.DataFrame()

    data_d = yf.download(
        tickers,
        period="10d",     # holgura por feriados/fines de semana
        interval="1d",
        group_by="ticker",
        auto_adjust=False,
        threads=True,
        progress=False
    )

    rows = []
    for tk in tickers:
        try:
            df = data_d[tk].copy() if isinstance(data_d.columns, pd.MultiIndex) else data_d.copy()
        except Exception:
            df = data_d.copy()

        if df.empty:
            print(f"[WARN] Sin datos diarios para {tk}")
            continue

        df = df.dropna(how="all")
        df.index = pd.to_datetime(df.index).tz_localize(None)
        df = df.sort_index()

        if "Close" not in df.columns:
            print(f"[WARN] Columna Close faltante en {tk}")
            continue

        closes = df["Close"].dropna()
        if closes.shape[0] < 2:
            print(f"[WARN] Menos de 2 cierres válidos para {tk}")
            continue

        close_today = closes.iloc[-1]
        close_yday  = closes.iloc[-2]

        if pd.isna(close_today) or pd.isna(close_yday) or close_yday == 0:
            print(f"[WARN] Datos insuficientes para rendimiento en {tk}")
            continue

        ret_abs = close_today - close_yday
        ret_pct = (close_today / close_yday - 1.0) * 100.0

        rows.append({
            "ticker": tk,
            "close_yday": close_yday,
            "close_today": close_today,
            "ret_abs": ret_abs,
            "ret_pct": ret_pct
        })

    return pd.DataFrame(rows)

def to_latex_table_daily(df: pd.DataFrame, title: str, label: str) -> str:
    """
    Convierte ['ticker','close_today','ret_abs','ret_pct'] a tabla LaTeX (booktabs).
    """
    if df is None or df.empty:
        return ("% Tabla vacía: " + title + "\n")

    df = df.copy()
    for c in ["close_today","ret_abs","ret_pct","close_yday"]:
        if c in df.columns:
            df[c] = pd.to_numeric(df[c], errors="coerce")

    if "ticker" in df.columns:
        df["ticker"] = df["ticker"].astype(str).apply(latex_escape)

    pretty = df.rename(columns={
        "ticker": "Emisora",
        "close_today": "Precio",
        "ret_abs": r"$\Delta$ Abs.",
        "ret_pct": r"$\Delta$ Día (\%)"
    })[["Emisora","Precio",r"$\Delta$ Abs.",r"$\Delta$ Día (\%)"]]

    pretty = pretty.round({
        "Precio": 4,
        r"$\Delta$ Abs.": 4,
        r"$\Delta$ Día (\%)": 2
    })

    header = (
        "\\begin{table}[htbp]\n"
        "\\centering\n"
        "\\footnotesize\n"
        f"\\caption{{{title}}}\n"
        f"\\label{{{label}}}\n"
        "\\begin{tabular}{lrrr}\n"
        "\\toprule\n"
        "Emisora & Precio & $\\Delta$ Abs. & $\\Delta$ Día (\\%) \\\\\n"
        "\\midrule\n"
    )

    body_lines = []
    for _, row in pretty.iterrows():
        line = (
            f"{row['Emisora']} & "
            f"{row['Precio']:.4f} & "
            f"{row['$\\Delta$ Abs.']:.4f} & "
            f"{row['$\\Delta$ Día (\\%)']:.2f}\\% \\\\"
        )
        body_lines.append(line)

    footer = "\\bottomrule\n\\end{tabular}\n\\end{table}\n"
    return header + "\n".join(body_lines) + "\n" + footer


# =========================================================
# --- Cálculo: rendimiento diario + ranking Best/Worst ---
# =========================================================
ret_df = get_daily_returns(TICKERS)

if ret_df.empty:
    print("\n[WARN] No hay suficientes datos para ranking. Revisa símbolos o reintenta más tarde.")
    top_gainers = pd.DataFrame()
    top_losers  = pd.DataFrame()
else:
    top_gainers = ret_df.nlargest(5, "ret_pct").copy()
    top_losers  = ret_df.nsmallest(5, "ret_pct").copy()

# =========================================================
# --- LaTeX: generar tablas ---
# =========================================================
latex_gainers = to_latex_table_daily(
    top_gainers,
    "Top 5 Ganadoras del S\\&P/BMV IPC",
    "tab:ipc_top5_gainers"
)
latex_losers  = to_latex_table_daily(
    top_losers,
    "Top 5 Perdedoras del S\\&P/BMV IPC",
    "tab:ipc_top5_losers"
)

print("=== Copia y pega en Overleaf ===\n")
print(latex_gainers)
print("\n")
print(latex_losers)

# Guardar a archivo .tex (opcional)
with open("/content/top_movers.tex", "w", encoding="utf-8") as f:
    f.write(latex_gainers + "\n\n" + latex_losers)

print("\nArchivo guardado: /content/top_movers.tex")
```
